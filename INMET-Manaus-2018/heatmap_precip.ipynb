```python
import pandas  as pd 
import numpy   as np
import seaborn as sns
import calmap  as cm

import matplotlib.pyplot as plt
```

```python
sns.set(style='ticks')
```

## Abrir arquivo

```python
df = pd.read_csv('MAO2018INMET.csv')
```

```python
datas = pd.to_datetime(df['data'], format='%d/%m/%Y')
```

```python
dia = pd.DatetimeIndex(datas).day
mes = pd.DatetimeIndex(datas).month
```

```python
smn = pd.DatetimeIndex(datas).dayofweek
```

## Acumulado Diário

```python
acmdiario = df.groupby([datas]) # agrupar por todos os dias
```

```python
acmdiario = acmdiario.sum() # acumulado diário
```

```python
prpdiaria = acmdiario['precipitacao']
```

## Heatmap

```python
sns.set_context('poster')
```

```python
plt.figure(figsize=(15,5))

cm.yearplot(prpdiaria, year=2018,
            cmap='BuPu', vmin=0.1, vmax=30, fillcolor='White',
           linecolor='Grey',
           daylabels=['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sab', 'Dom'], dayticks=True,
           monthlabels=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul',
                        'Ago', 'Set', 'Out', 'Nov', 'Dez'], monthticks=True)

plt.title('Dias com Precipitação em Manaus/AM em 2018')

plt.tight_layout()
plt.savefig('prec.png')
plt.show()
```
