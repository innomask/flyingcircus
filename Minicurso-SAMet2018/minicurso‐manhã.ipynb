#### Exemplo 1: mas o que é um array?

Um array é basicamente uma lista de valores, como um conjunto de dados ou uma lista de compras do supermercado, interpretada diretamente pelo NumPy.

```python
# importar Numpy
import numpy  as  np
```

```python
array1 = np.array([1,2,3,4,5])
```

Acima criamos um array com valores de 1 a 5, note que é preciso usar a função "array" do pacote NumPy.

```python
print(array1)
```

O 'array1' pode ser visto como um vetor em linha simples, mas e se for preciso criar uma matriz?

```python
matriz1  =  np.array( [ [1,2], [3,4] ])
```

```python
print(matriz1)
```

Vejamos as dimensões dos arrays que acabamos de criar:

```python
print(array1.shape, matriz1.shape)
```

O 'array1' é um vetor linha com 5 elementos, enquanto que 'matriz1' é uma matriz 2x2.

#### Exemplo 3: uma outra forma de criar arrays.

Uma maneira mais automática de criar um array é através da função "np.arange()".

```python
array2 = np.arange(0,10)
print(array2)
```

Note que o array 2 foi 0 a 9, e não de 0 a 10. Mas por que isso acontece?

A estrutura do "np.arange()" é a seguinte:

np.arange(valor inicial, valor final - passo de tempo, passo de tempo)

```python
np.arange(0,11,1)
```

```python
np.arange(0,12,2)
```

Basicamente, np.arange() nos dá a possibilidade de criar um array qualquer a partir de um loop sem ser preciso criar um loop.

#### Exemplo 2: função de Planck

$ B_{\lambda}(T): \frac {C_{1} \lambda^{-5}} {\pi (exp( \frac{C_{2}} {\lambda T}) -1)} $, onde $C_{1} = 3.74 \times 10^{-16} W m^{-2} $ and $C_{2} = 1.45 \times 10^{-2} m K$.

```python
# definir comprimentos de onda
wvlg = np.arange(0.1,2.0,0.01) # wavelength
lmbd = wvlg * 1E-6
```

```python
C1  =  3.74E-16
C2  =  1.45E-2
T1  =  5000.    # temperatura absoluta (kelvin)
Ca  =  C2 / (lmbd * T1)
```

```python
Pa  =  C1 * (lmbd ** -5.) / (np.pi * ( np.exp(Ca)-1) )
```

```python
plt.plot(wvlg, Pa * 1E-6)
plt.show()
```

Vamos fazer a nossa própria versão da figura clássica da função de Planck.

```python
T2  =  6000 
T3  =  7000

Cb  =  C2 / (lmbd * T2)
Cc  =  C2 / (lmbd * T3)

Pb  =  C1 * (lmbd ** -5.) / (np.pi * (np.exp (Cb) -1))
Pc  =  C1 * (lmbd ** -5.) / (np.pi * (np.exp (Cc) -1))
```

```python
plt.plot(wvlg, Pa * 1E-6,'blue',   label='5000 K')
plt.plot(wvlg, Pb * 1E-6,'orange', label='6000 K')
plt.plot(wvlg, Pc * 1E-6,'red',    label='7000 K')

plt.legend(loc='upper right')

plt.title('Função de Planck')
plt.xlabel(r'Comprimento de Onda ($\mu m$)') # sim, dá pra usar Latex!
plt.ylabel(r'$ B_{\lambda} (W m^{2} sr^{-1} \mu m) \times 10^{7}  $')


plt.show()
```

#### Exemplo 3: plots em 3D

[Exemplo da biblioteca do Matplotlib](https://matplotlib.org/gallery/pyplots/whats_new_99_mplot3d.html#sphx-glr-gallery-pyplots-whats-new-99-mplot3d-py)

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, cmap=cm.viridis)

plt.show()
```

#### Exemplo 4: o uso de subplots

```python
x = np.arange(0.0, 10, 0.1)
cosseno = np.cos(x)
seno    = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout=True)
print(axs.shape)

axs[0].plot(x, seno)
axs[0].set_xlabel('Tempo')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Função Seno')

axs[1].plot(x, cosseno)
axs[1].set_xlabel('Tempo')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Função Cosseno')


plt.show()
```
